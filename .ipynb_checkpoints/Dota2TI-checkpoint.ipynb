# Exploratory Analysis on Dota2 The International Tournaments from 2011 to 2021

### What is Dota2?

Dota 2 is a popular online multiplayer video game developed and published by Valve Corporation. It is a free-to-play game, and its gameplay involves two teams of five players each, who battle against each other in a large map called the "map of the Ancients."

Each player controls a hero character with unique abilities and strengths, and they must work together to destroy the enemy team's "Ancient," a large structure located at the heart of their base. The game is highly competitive and strategic, with players needing to coordinate their actions, manage resources, and make quick decisions in order to outwit and defeat their opponents.

Dota 2 has a large and active player base, with many professional esports teams and tournaments dedicated to the game. It has become one of the most popular and influential video games of all time, with millions of players around the world.

### What is The International Tournament?

The International Dota 2 Tournament is an annual esports event organized by Valve Corporation, the developer of Dota 2. It is considered one of the most prestigious and high-profile events in the esports industry.

The tournament features the best Dota 2 teams from around the world competing against each other for a multi-million dollar prize pool. The tournament usually takes place in August or September and lasts for several days.

The International Dota 2 Tournament has a unique format, where teams compete in a double elimination bracket until only two teams remain. The final match is a best-of-five series, with the winner being crowned the champion of the tournament.

The International has become famous for its enormous prize pools, which are funded in part by sales of in-game content. The prize pool for The International 2021 was over $40 million, making it one of the largest prize pools in the history of esports.

## Data Preparation and Cleaning

The data was taken from kaggle.
Link: https://www.kaggle.com/datasets/arpan129/dota-2-the-international-complete-dataset

The small size of the data allowed the author to download the data in order to work with the project offline.

Note: Due to COVID-19, The International 2020 is cancelled

```python
import pandas as pd
import difflib
import numpy as np
```

Create a list of folders

```python
folder_list = []
for i in range(11,22):
    folder_list.append(f"TI_20{i}")

folder_list
```

Creating the Prize Pool Dataframe

```python
prize_pool_year = []
prize_pool_value = []
def prizepoolfunc(folder):
    data = pd.read_csv(f'{folder}/PrizePoolFinalStandings.csv')
    #Column names are not uniform, I used index location in order to get around this problem 
    data['Prize'] = [int(x.replace("$","").replace(",","").replace("-","0")) for x in data.iloc[:,2]]
    prize_pool_year.append(folder)
    prize_value = data['Prize'].sum()
    prize_pool_value.append(prize_value)
#Since The International 2022 is cancelled, we don't want to include it

for i in range(0,len(folder_list)):
    try:
        prizepoolfunc(folder_list[i])
    except:
        pass

prize_pool = pd.DataFrame({'Year':prize_pool_year,
              'Value':prize_pool_value})

prize_pool['Value'] = prize_pool['Value'].map('{:,.0f}'.format)

prize_pool
```

Create the Country Representation Dataframe

```python
#We need a distinct list of countries
country_set = set()
for i in range(0,len(folder_list)):
    try:
        data = pd.read_csv(f"{folder_list[i]}/Country Representation.csv")
        country_set = country_set.union(set(data.iloc[:,1]))
    except:
        pass
'''
There are duplicate countries that needed to be removed
In order to get around this later, we will need to use fuzzy matching
'''
country_set.remove('United States of America')
country_set.remove('Bosnia and Herzegovina')
country_list = list(country_set)
country_list.sort()

country_rep = pd.DataFrame({"Country":country_list})
for i in range(0,len(folder_list)):
    try:
        right_data = pd.read_csv(f"{folder_list[i]}/Country Representation.csv")
        right_data = right_data.iloc[:,1:3]
        right_data = right_data.rename(columns={"No. of Players":f"{folder_list[i]}"})
        right_data['Country'] = right_data['Country'].apply(lambda x: difflib.get_close_matches(x, country_rep['Country'])[0])
        country_rep = country_rep.merge(right_data, how="left", left_on="Country", right_on="Country")
    except:
        pass

country_rep[['TI_2021','remove_me']] = country_rep['Representation'].str.split("/", expand=True)
country_rep = country_rep[['Country', 'TI_2011', 'TI_2012', 'TI_2013', 'TI_2014', 'TI_2015',
       'TI_2016', 'TI_2017', 'TI_2018', 'TI_2019', 
       #'Representation', 
       'TI_2021',
       #'remove_me'
       ]]

country_rep = country_rep.replace(np.nan,0)

for i in range (0,len(folder_list)):
    try:
        country_rep = country_rep.astype({f'{folder_list[i]}':'int32'})
    except:
        pass

country_rep
```

Create Hero Statistics Dataframe

Note: Due to incosistency of the dataset, the dataframe will only include the data starting from 2016 onwards

```python
#Create a new list of folders
new_list = folder_list[5:]
#Create a distinct list of heroes
hero_set = set()
for i in range(0,len(new_list)):
    try:
        data = pd.read_csv(f"{new_list[i]}/Statistics/Hero Statistics.csv")
        hero_set = hero_set.union(set(data.iloc[:,1]))
    except:
        pass

hero_list = list(hero_set)
hero_list.sort()
hero_df = pd.DataFrame({'Hero':hero_list})
hero_df
```

```python
#update folder list
new_list.pop(4)
folder_list = new_list
folder_list
```

```python
#We need the following dataframes that is related to our hero dataframe
h_times_picked = hero_df 
h_pick_perc = hero_df
hr_times_pick = hero_df
hr_win_perc = hero_df
hd_times_pick = hero_df
hd_win_perc = hero_df
hero_index = 1
column_index = [2,6,7,10,11,14]
df_list = ['h_times_picked', 'h_pick_perc', 'hr_times_pick', 'hr_win_perc', 'hd_times_pick', 'hd_win_perc']

def herofunc(funcfolder,funcindex,funcdataframe):
    right_data = pd.read_csv(f'{funcfolder}/Statistics/Hero Statistics.csv')
    right_data = right_data.iloc[:,[hero_index,funcindex]]
    column_to_change = right_data.columns[1]
    right_data = right_data.rename(columns={column_to_change: funcfolder})
    funcdataframe = funcdataframe.merge(right_data, how='left', left_on='Hero', right_on='Hero')
    funcdataframe = funcdataframe.replace(np.nan,0)
    return funcdataframe
    
folder = 0
while folder < len(folder_list):
    h_times_picked = herofunc(f'{folder_list[folder]}', 2, h_times_picked)
    h_pick_perc = herofunc(f'{folder_list[folder]}', 6, h_pick_perc)
    hr_times_pick = herofunc(f'{folder_list[folder]}', 7, hr_times_pick)
    hr_win_perc = herofunc(f'{folder_list[folder]}', 10, hr_win_perc)
    hd_times_pick = herofunc(f'{folder_list[folder]}', 11, hd_times_pick)
    hd_win_perc = herofunc(f'{folder_list[folder]}', 14, hd_win_perc)
    folder = folder + 1

def toChangetype(d,t):
    return d.astype({'TI_2016':t,'TI_2017':t,'TI_2018':t,'TI_2019':t,'TI_2021':t})

h_times_picked = toChangetype(h_times_picked,int)
hr_times_pick = toChangetype(hr_times_pick,int)
hd_times_pick = toChangetype(hd_times_pick,int)
```

Change percentages to numeric:

```python
def toCorrect(d,v):
    return d[v].str.rstrip('%').replace([np.nan,"-"],0)

modify = [h_pick_perc,hr_win_perc,hd_win_perc]

for  x in modify:
    for i in folder_list:
        x[i] = toCorrect(x,i)

h_pick_perc = toChangetype(h_pick_perc,float)
hr_win_perc = toChangetype(hr_win_perc,float)
hd_win_perc = toChangetype(hd_win_perc,float)
```

Create totals/average column

```python
def createColumn(d,action):
    if action =='sum':
        d['Total'] = d['TI_2016'] + d['TI_2017'] + d['TI_2018'] + d['TI_2019'] + d['TI_2021']
    elif action =='average':
        d['Average'] = (d['TI_2016'] + d['TI_2017'] + d['TI_2018'] + d['TI_2019'] + d['TI_2021'])/5
    else:
        pass

modify2 = [h_times_picked,hr_times_pick,hd_times_pick]

for i in modify2:
    i = createColumn(i,'sum')

modify3 = [h_pick_perc,hr_win_perc,hd_win_perc]
for i in modify3:
    i = createColumn(i,'average')
```

*Question: Why is it that the price pool for the Dota2 The International Tournament increasing year by year?*
1. Crowdfunding: The International's prize pool is largely crowdfunded through the sale of in-game items and Battle Passes. Each year, Valve, the developer of Dota 2, releases a Battle Pass that allows players to purchase cosmetic items, access exclusive features, and contribute to the prize pool. A percentage of the revenue generated from these sales goes towards the prize pool, and as the Dota 2 player base has grown over the years, so has the amount of money raised.
2. Player and Community Support: Dota 2 has a passionate and dedicated player community. Many players and fans eagerly contribute to the prize pool to support the competitive scene and their favorite teams. The desire to see the prize pool reach new heights has become somewhat of a tradition and a point of pride within the Dota 2 community.
3. Stretch Goals and Rewards: Valve has been creative in incentivizing players to contribute more to the prize pool. They introduce stretch goals, which are additional rewards or features unlocked when certain funding milestones are reached. These stretch goals can include exclusive in-game items, cosmetic effects, and even new game modes. The introduction of these enticing rewards motivates players to spend more on the Battle Pass, thus driving up the prize pool.
4. Prestige and Attention: The International is widely regarded as the most prestigious tournament in Dota 2, attracting the best teams from around the world. The ever-increasing prize pool adds to the tournament's allure and raises its significance in the esports scene. The attention generated by the escalating prize pool helps in attracting more viewers, sponsors, and media coverage, thereby further increasing the prize pool.
5. Media and Streaming Opportunities: Esports has gained mainstream recognition, and Dota 2 is a prominent title in the competitive gaming industry. The International offers significant media exposure and attracts millions of viewers through online streaming platforms. This increased viewership translates to higher advertising and sponsorship revenue, which can be reinvested into the prize pool.
These factors, combined with the growing popularity of Dota 2 and the dedication of its community, have contributed to the year-on-year increase in The International's prize pool.

Country Representation Charts:

```python
folder_list = list(country_rep.columns)
folder_list.pop(0)

fig, axes = plt.subplots(5,2, figsize=(15,30))
def createPie(x):
    ax.set_title(f'{x} Country Representative')
    top_5 = list(country_rep.sort_values(x, ascending=False).head(5)['Country'])
    top_5.append('Others')
    top_5_count = list(country_rep.sort_values(x, ascending=False).head(5)[x])
    others_count = country_rep.sort_values(x,ascending=False)
    others_count = others_count.iloc[5:]
    others_count = others_count[x].sum()
    top_5_count.append(others_count)
    Country = top_5
    player_count =  top_5_count
    ax.pie(player_count, labels=Country, autopct='%.2f%%', startangle=90)

folder_index = 0
for r in range(0,5):
    for c in range(0,2):
        ax = axes[r][c]
        createPie(folder_list[folder_index])
        folder_index  = folder_index + 1
```

*Observation*

On all the tournaments, the majority of the players came from China.


*Question: Why is it that majority of Professional Dota2 Players came from China?*

1. Player Base and Popularity: China has a massive player base for Dota 2, with a large number of players actively participating in the game. This vast pool of players provides a strong foundation for talent development. The popularity of Dota 2 in China creates a competitive environment that encourages players to strive for excellence and invest significant time and effort into honing their skills.
2. Infrastructure and Support: China has established a robust esports infrastructure with numerous professional organizations, teams, and training facilities dedicated to Dota 2. These organizations provide support, resources, and infrastructure necessary for aspiring players to pursue a career in esports. There are also well-established leagues and tournaments within China, offering ample opportunities for players to showcase their skills and attract attention from top teams.
3. Government Support: The Chinese government has recognized the potential of esports and has actively supported its development. They have invested in esports infrastructure, established training centers, and provided financial incentives to promote the growth of the industry. This support has created a favorable environment for Dota 2 players in China to thrive and succeed.
4. Dota 2 Culture and Legacy: Dota, the precursor to Dota 2, originated in China. Chinese players have a long-standing history and tradition in the game, with many skilled players emerging from the region since the early days of Dota. This legacy has influenced the development of Dota 2 in China and fostered a strong competitive spirit and dedication to the game.
5. Work Ethic and Discipline: Chinese players are often known for their strong work ethic, discipline, and dedication to improvement. They are known to put in long hours of practice, study replays, and analyze strategies to enhance their gameplay. This commitment to continuous improvement has played a significant role in the success of Chinese Dota 2 players.
6. Regional Competitiveness: China has a highly competitive Dota 2 scene, with numerous strong teams and players vying for success. The intense competition within the region pushes players to elevate their skills and strategies. The presence of strong opponents provides valuable practice and learning opportunities, further fueling the growth of Chinese Dota 2 talent.


While China has produced many top-tier Dota 2 players, it is important to note that the Dota 2 scene is global, with skilled players hailing from various regions. The success of Chinese players is a combination of the factors mentioned above, but talent can emerge from any part of the world given the right circumstances and opportunities.

Radiant Pick Rate Win Rate Scatter Plot:



```python
hr_times_pick = hr_times_pick[['Hero','Total']]
hr_win_perc = hr_win_perc[['Hero','Average']]
radiant = pd.merge(hr_times_pick, hr_win_perc, how='left', left_on='Hero', right_on='Hero')
radiant = radiant.rename(columns={'Total':'Total Picks', 'Average': 'Average Winrate(%)'})

sns.scatterplot(x='Average Winrate(%)', y='Total Picks', data=radiant).set(title='Radiant Pick Total/Winrate Average Ratio')
```

*Observation:*


It seems like the data has a high Coefficient of Determination. Also, the number of dots on the around along the area between 25% and 60% winrate is larger compared to the areas around them. 

```python
rhist = radiant['Average Winrate(%)']

plt.hist(rhist, bins=20)
plt.xlabel('Average Winrate(%)')
plt.ylabel('Frequency')
plt.title('Radiant Winrate Frequency')
plt.show()
```

*Observation:*


This histogram confirmed the author's observation on the scatterplot.

Dire Pick Rate Win Rate Scatterplot:


```python
hd_times_pick = hd_times_pick[['Hero','Total']]
hd_win_perc = hd_win_perc[['Hero','Average']]
dire = pd.merge(hd_times_pick, hd_win_perc, how='left', left_on='Hero', right_on='Hero')
dire = dire.rename(columns={'Total':'Total Picks', 'Average': 'Average Winrate(%)'})

sns.scatterplot(x='Average Winrate(%)', y='Total Picks', data=dire).set(title='Dire Pick Total/Winrate Average Ratio')
```
